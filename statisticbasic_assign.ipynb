1. What is Statistics, and Why is it Important?


ans- Statistics is the study of collecting, organizing, analyzing, interpreting, and presenting data.
Importance: It helps in decision-making, understanding patterns, and predicting future outcomes in fields like economics, science, business, and healthcare.

2. What are the Two Main Types of Statistics?


ans-Descriptive Statistics: Summarizes or describes the characteristics of a dataset.
Inferential Statistics: Makes predictions or inferences about a population using sample data.


3. What are Descriptive Statistics?


ans-Descriptive statistics involve numerical measures such as:
a- Mean, Median, Mode
b- Range, Variance, Standard Deviation
c- Charts & Graphs like histograms and pie charts



4. What is Inferential Statistics?



ans- Involves drawing conclusions about a population based on a sample.
Common techniques include: Hypothesis Testing, Confidence Intervals, Regression Analysis



5. What is Sampling in Statistics?


ans-Sampling is the process of selecting a subset (sample) from a larger population to make statistical inferences.



6. What are the Different Types of Sampling Methods?



ans-Random Sampling, Systematic Sampling, Stratified Sampling, Cluster Sampling, Convenience Sampling




7. What is the Difference Between Random and Non-Random Sampling?




ans- Random Sampling: Every individual has an equal chance of selection.
     Non-Random Sampling: Based on convenience or judgment; may introduce bias.




8. Define and Give Examples of Qualitative and Quantitative Data



ans-Qualitative Data (Categorical): Describes qualities or characteristics.
Example: Gender, color, nationality.

Quantitative Data (Numerical): Expressed as numbers.
Example: Age, height, income.




9. What are the Different Types of Data in Statistics?



ans-Nominal: Categories without order (e.g., colors).
    Ordinal: Categories with order (e.g., rankings).
    Interval: Ordered with equal intervals, no true zero (e.g., temperature).
    Ratio: Ordered with equal intervals and a true zero (e.g., weight).




10. Explain Nominal, Ordinal, Interval, and Ratio Levels of Measurement



ans-Nominal: Names or labels (no rank).
    Ordinal: Ordered categories (ranked, but no consistent difference).
    Interval: Ordered, equal spacing, no absolute zero.
    Ratio: Interval level with a meaningful zero, allowing true ratios.




11. What is the Measure of Central Tendency?



ans-Describes the center or average of a dataset. Includes mean, median, and mode.




12. Define Mean, Median, and Mode



ans-Mean: The average of all values.
    Median: The middle value when data is ordered.
    Mode: The most frequently occurring value.




13. What is the Significance of the Measure of Central Tendency?



ans-Helps understand the general trend of the data. Provides a single value to represent the dataset.

15. What is Standard Deviation, and Why is it Important?



ans-Standard Deviation is the square root of variance. It tells how much the data deviates from the mean. it is More interpretable because it's in the same units as the data.




16. Define and Explain the Term Range in Statistics



ans-Range = Maximum value - Minimum value
Indicates the spread or dispersion of the dataset.



17. What is the Difference Between Variance and Standard Deviation?



ans-Variance is in squared units.
Standard Deviation is in the same units as the data, making it more interpretable.



18. What is Skewness in a Dataset?



ans-Measures the asymmetry of the distribution.
    Positive skew: Tail is on the right.
    Negative skew: Tail is on the left.




19. What Does it Mean if a Dataset is Positively or Negatively Skewed?




ans-Positively Skewed: Mean > Median > Mode; long right tail.
    Negatively Skewed: Mean < Median < Mode; long left tail.




20. Define and Explain Kurtosis



ans-Measures the "tailedness" or peak of a distribution.
    High kurtosis: Heavy tails, more outliers.
    Low kurtosis: Light tails, fewer outliers.



21. What is the Purpose of Covariance?



ans-Measures how two variables change together.
    Positive covariance: variables increase together.
    Negative covariance: one increases, the other decreases.



22. What Does Correlation Measure in Statistics?



ans-Measures the strength and direction of a linear relationship between two variables.

Ranges from -1 to +1:
    +1: Perfect positive correlation
    0: No correlation
   -1: Perfect negative correlation



23. What is the Difference Between Covariance and Correlation?



ans-Covariance: Unstandardized, only shows direction.
    Correlation: Standardized, shows both strength and direction.




24. What Are Some Real-World Applications of Statistics?




ans-Business: Market analysis, forecasting.
    Medicine: Drug testing, medical studies.
    Government: Census, policy planning.
    Sports: Performance analysis.
    Education: Test scoring, curriculum effectiveness.

```python
#4. Implement sampling techniques like random sampling and stratified sampling

import pandas as pd

df = pd.DataFrame({
    'Group': ['A']*5 + ['B']*5,
    'Score': np.random.randint(60, 100, 10)
})

# Random sampling
random_sample = df.sample(n=4)

# Stratified sampling
strat_sample = df.groupby('Group', group_keys=False).apply(lambda x: x.sample(2))

print("Random Sample:\n", random_sample)
print("Stratified Sample:\n", strat_sample)
```

```python
#5. Write a Python function to calculate the range of a dataset

def calc_range(data):
    return max(data) - min(data)

print("Range:", calc_range(data))
```

```python
#13. Calculate the mean, median, and mode of grouped data

grouped = pd.DataFrame({
    'Class': ['0-10', '10-20', '20-30'],
    'Frequency': [5, 8, 7]
})

grouped['Midpoint'] = [5, 15, 25]
grouped['fx'] = grouped['Frequency'] * grouped['Midpoint']

grouped_mean = grouped['fx'].sum() / grouped['Frequency'].sum()
print("Grouped Mean:", grouped_mean)
```

```python
#15. Use NumPy or pandas to summarize a datasetâ€™s descriptive statistics

import pandas as pd
import numpy as np

data = np.random.normal(loc=50, scale=10, size=1000)
df = pd.DataFrame(data, columns=['Scores'])

# Summary statistics
print(df.describe())
```

```python
#16. Plot a boxplot to understand the spread and identify outliers

import matplotlib.pyplot as plt

plt.boxplot(df['Scores'])
plt.title("Boxplot of Scores")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```

```python
#17. Calculate the interquartile range (IQR) of a dataset

Q1 = np.percentile(df['Scores'], 25)
Q3 = np.percentile(df['Scores'], 75)
IQR = Q3 - Q1

print("Interquartile Range (IQR):", IQR)
```

```python
#18. Implement Z-score normalization and explain its significance

z_scores = (df['Scores'] - df['Scores'].mean()) / df['Scores'].std()
df['Z-Score'] = z_scores

print(df.head())
# Significance:
#Z-score normalization standardizes data, making it easier to compare features with different units or scales. It helps in outlier detection and is commonly used in machine learning preprocessing.

```
